```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


```

```python
# Load Data
data = pd.read_csv("Iris.csv", index_col="Id")
X = data.drop("Species", axis=1)
y = pd.factorize(data['Species'])[0]

```

```python
# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

```

```python
# Plot
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap="viridis")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.title("t-SNE on Iris Dataset")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Load dataset
data = pd.read_csv("Iris.csv", index_col="Id")
X = data.drop("Species", axis=1)


```

```python

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# DBSCAN model
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Add labels to dataset
data["DBSCAN_Cluster"] = labels
```

```python

# Scatter plot visualization
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels, cmap="plasma", s=50)
plt.xlabel("Standardized SepalLengthCm")
plt.ylabel("Standardized SepalWidthCm")
plt.title("DBSCAN Clustering on Iris Dataset")
plt.colorbar(label="Cluster ID")
plt.show()

# Check cluster distribution
print(data["DBSCAN_Cluster"].value_counts())
```
