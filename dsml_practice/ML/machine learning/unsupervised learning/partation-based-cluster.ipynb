```python
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('Iris.csv',index_col='Id')
```

```python
data.head()
```

```python
t,_ = pd.factorize(data.Species)
data.plot.scatter(x='SepalLengthCm', y='PetalLengthCm', c=t, cmap="viridis")
```

```python
X= data.drop('Species', axis=1).to_numpy()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
kmeans_model= KMeans(n_clusters=3, init= 'k-means++', max_iter=200, random_state=200)
```

```python
kmeans_model.fit(x_scaled)
```

```python
kmeans_model.cluster_centers_
```

```python
kl=kmeans_model.labels_
```

```python
data.plot.scatter(x='SepalLengthCm',y='PetalLengthCm',c=kl, cmap='viridis')
```
