Numerical columns → Median imputation

Why not mean?
Numerical variables in housing data (e.g., LotFrontage, SalePrice) often have skewed distributions and outliers (very large or very small lots).

Using the mean would be influenced by those outliers.

Using the median is more robust because it represents the central tendency without being distorted.

Example: For LotFrontage, extreme large values would pull the mean upward, but the median remains stable.
