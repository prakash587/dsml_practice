```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2️⃣ Scatter Plot: Sepal Length vs Petal Length, colored by Species
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=iris_df,
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species',
    palette='Set2'
)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Species')
plt.show()
```

```python
# 3. Box Plot of Sepal Width grouped by Species
plt.figure(figsize=(6,4))
sns.boxplot(data=iris_df, x="Species", y="SepalWidthCm", palette="Set3", hue = None)
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```


7. **Correlation Matrix**
   - Calculate the correlation matrix for the numerical features and create a heatmap to visualize the correlations using `seaborn`.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Select only numerical columns
numerical_cols = iris_df.select_dtypes(include='number')

# 2️⃣ Compute the correlation matrix
corr_matrix = numerical_cols.corr()


```

```python
# 3️⃣ Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,
    annot=True,        # show correlation values
    cmap='coolwarm',   # color map
    linewidths=0.5
)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
