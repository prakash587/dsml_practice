```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 1. Create dataset
data = {
    "Age": ["<30", "<30", "31-40", ">40", ">40", ">40", "31-40", "<30", "<30", ">40", "<30", "31-40", "31-40", ">40"],
    "Income": ["High", "High", "High", "Medium", "Low", "Low", "Low", "Medium", "Low", "Medium", "Medium", "Medium", "High", "Medium"],
    "Student": ["No", "No", "No", "No", "Yes", "Yes", "Yes", "No", "Yes", "Yes", "Yes", "No", "Yes", "No"],
    "Credit_Rating": ["Fair", "Excellent", "Fair", "Fair", "Fair", "Excellent", "Excellent", "Fair", "Fair", "Fair", "Excellent", "Excellent", "Fair", "Excellent"],
    "Buys_Computer": ["No", "No", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes", "Yes", "Yes", "Yes", "Yes", "No"]
}

```

```python
df = pd.DataFrame(data)

```

```python
# 2. Convert categorical variables to numeric (one-hot encoding)
X = pd.get_dummies(df.drop("Buys_Computer", axis=1))
y = df["Buys_Computer"]

```

```python
# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# 4. Train decision tree
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=42)
clf.fit(X_train, y_train)

```

```python
# 5. Evaluate
print("Test Accuracy:", clf.score(X_test, y_test))

```

```python
# 6. Visualize
plt.figure(figsize=(14, 7))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()
```

```python
import numpy as np

# Get feature importances from the trained tree
importances = clf.feature_importances_

# Sort features by importance
indices = np.argsort(importances)[::-1]

print("Feature Importance Ranking:")
for i in indices:
    print(f"{X.columns[i]}: {importances[i]:.3f}")


```

```python
# Optional: plot them
plt.figure(figsize=(10,5))
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=45)
plt.title("Decision Tree Feature Importance")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

for depth in [2, 3, 4, None]:  # None = no limit
    clf = DecisionTreeClassifier(criterion="entropy", max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    
    train_acc = accuracy_score(y_train, clf.predict(X_train))
    test_acc = accuracy_score(y_test, clf.predict(X_test))
    
    print(f"Max Depth={depth}: Train Acc={train_acc:.2f}, Test Acc={test_acc:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    "criterion": ["gini", "entropy"],
    "max_depth": [2, 3, 4, 5, None],
    "min_samples_split": [2, 3, 4],
    "min_samples_leaf": [1, 2, 3]
}

```

```python
# Set up GridSearch with cross-validation
grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring="accuracy"
)
```

```python

# Fit the model
grid_search.fit(X_train, y_train)

```

```python
# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best CV Accuracy:", grid_search.best_score_)

# Evaluate on test set
best_clf = grid_search.best_estimator_
test_acc = best_clf.score(X_test, y_test)
print("Test Accuracy with Best Model:", test_acc)
```
